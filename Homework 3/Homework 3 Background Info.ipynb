### <font color='blue'>Nested loops </font> 

In one of the exercises in hw 3 you will need to generate a list of <tt>ns</tt> numbers, each of which is a sum of <tt>nr</tt> pseudo-random numbers distributed uniformly in the interval $[0,1]$. This can be done using nested loops - a nested loop is a loop contained within other (loops), this is indicated (as always in Python) with proper indentation

```python
#  ns sums 
ns, nr = 5, 3 # you will want much larger numbers here for ns, and experiment with different nr
for i in range(ns): # i will change from 0 to ns-1
    # compute a sum on nr random numbers
    for j in range(nr): # j will change from 0 to nr-1
        # sum nr random numbers here, i is constant while this loop is running
        print(i, j)
```

### <font size=4 color='blue'>Plotting histograms with Matplotlib</font> 

[Matplotlib](https://matplotlib.org/) is the main plotting module in Python. We will be using it throughout the rest of the course, introducing different features as the need arises. If you want to explore now, you can find various Matplotlib tutorials [here](https://matplotlib.org/stable/tutorials/index.html). 

In this assignment you are asked to plot a histogram (or "bar-chart") of values of the sums of <tt>nr</tt> random numbers.
Example code below shows how a histogram for the values of random numbers themselves can be plotted. You will simply need to substitute your sum values into the <tt>xs</tt> list used below. 

The code in the cell below imports function <tt>random</tt> from Python module called <tt>random</tt> (somewhat confusingly module and function have the same name. Function  <tt>random</tt>, when called, returns a single float (pseudo) random number uniformly distributed in $[0,1]$. 

```python
# generate nr random numbers 
from random import random 
xs = [] # initialize xs with an empty list
nr = 10000 # generate nr random numbers specified here
for i in range(nr): 
    xs.append(random())
```

```python
# import pyplot package from Matplotlib library
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6)) # define figure and its size
plt.xlabel('x') # x-axis label
plt.ylabel('density of samples') # y-axis label

# bins = 'auto' chooses number of bins roughly optimal for a a given sample size
# density = True normalizes histogram so that amplitude of all bars sums up to one
plt.hist(xs, density=True, color='slateblue', bins='auto') # produce histogram of xs values
plt.show() # show the entire plot
```

The histogram above is "flat" (does not have a pronounced peak). If we generate more and more random numbers it will look flatter and flatter (try it!). This is because the random numbers produced by the <tt>random()</tt> function are *uniformly distributed* in the interval $[0,1]$. In other words, we say that they are samples *drawn* from the uniform probability density function described in Section 3.3.3 of [David Morin's book](http://www.people.fas.harvard.edu/~djmorin/probability.pdf).

### Random number generators that can be used for tests of randomness in ex2  of hw 3

```python
def lcg_randu(xr, a=65539, m=2**31):
    '''
    Linear Congruential Generator of pseudo-random numbers
    Parameters: xr - integer, previous number in the sequence 
            a: integer, multiplier
            m: integer, period of the sequence
            a=65539, m=2**31 - are bad choices that were made in the infamous RANDU RNGs unleashed by IBM 50 yrs ago
    Returns: integer, the next number in the sequence 
    '''
    return (a*xr) % m
```

```python
def lcg1(xr, a=2937, m=2**31-1):
    '''
    Linear Congruential Generator of pseudo-random numbers
    Parameters: xr - integer, previous number in the sequence 
            a: integer, multiplier
            m: integer, period of the sequence
            a somewhat better, but still sub-optimal choice of a and m
    Returns: integer, the next number in the sequence 
    '''
    return (a*xr) % m

def lcg2(xr, a=1583458089, m=2**31-1):
    '''
    Linear Congruential Generator of pseudo-random numbers
    Parameters: xr - integer, previous number in the sequence 
            a: integer, multiplier
            m: integer, period of the sequence
            optimal choices for 32-bit integers a and m here from Table 2 of L'Ecuyer (1999) 
    Returns: integer, the next number in the sequence 
    '''
    return (a*xr) % m
```

```python
from random import randint
def random_randint(xr):
    return randint(0, 2**31-1)

import numpy as np
def numpy_randint(xr):
    return int(np.random.randint(0, 2**31-1, size=1))
```

All of these can be called in the same way:

```python
xr = randint(13343,1000000) # random initial seed

print(lcg_randu(xr), lcg1(xr), lcg2(xr), random_randint(xr), numpy_randint(xr))
```

You can set up your PRNG testing function to pass on specific PRNG. For example,

```python
def test_prng(xr, prng=lcg_randu):
    xr = prng(xr)
    return xr

# in this case test_prng will use lcg_random function because it is defined as default above    
print(test_prng(xr))

# in this case test_prng will use lcg1 function to print    
print(test_prng(xr, prng=lcg1))

# in this case test_prng will use lcg1 function to print    
print(test_prng(xr, prng=lcg2))

# in this case test_prng will use lcg1 function to print    
print(test_prng(xr, prng=random_randint))

# in this case test_prng will use lcg1 function to print    
print(test_prng(xr, prng=numpy_randint))
```
