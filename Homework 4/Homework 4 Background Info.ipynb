### <font color='darkblue'>String formatting

String formatting using Python is described in one of the materials assigned for reading (see [here](https://www.digitalocean.com/community/tutorials/how-to-use-string-formatters-in-python-3)). The parts that you may need in this homework are formatting of integers and floats <tt>{:d}</tt> ot <tt>{:.2f}</tt> and using <  or > for aligning numbers. In Exercise 4 you need to apply string formatting using <tt>{:s}</tt>. 

Opportunities to use string and integer formatting also exist in ex 2 and 3 if you display information in the plot title or legend. 


Formatting with string .format() method generally looks like this {n} or {:format}  or {n:format} in the location within the string where you want to insert a value or piece of a string.

    {n} where n is an integer will do no formatting, it just inserts the value of the n-th argument of format() in this location.

    {:format} will insert an element from format() argument corresponding to the order of this statement (if this is the 2nd insert, then second argument will be inserted, etc.). The value will be formatted as specified by format, which can be 

    format = ps for strings where p is the total number of positions allocated (e.g. 5s will allocated space for 5 character string)
    format = pd for integers where p is the total number of positions allocated
    format = p.qf where p is the total number of positions allocated for the number and q is the number of digits after the floating point to output

    {:>format} = when used with format with explicit p specified will align the output value to the right within the total number of positions allocated 
    {:<format} = when used with format with explicit p specified will align the output value to the right within the total number of positions allocated 
    {n:format} will insert n-th argument of format() into a given location in the string. 

```python
from math import pi, e

pie  = pi * e
ipie = 1 / pie

print('pie = {:s}'.format(str(pie)))
print('pie = {:>20s}'.format(str(pie)))
print('pie = {:>20d}'.format(int(pie)))
# notice that output below is integer but not the same as previous 
# here the value is rounded up, while above an integer part of float is output 
print('pie = {:>.0f}'.format(pie)) 
print('pie = {:>20.0f}'.format(pie)) 
print('pie = {:>7.3f}'.format(pie)) 
print('pie = {:<7.3f}'.format(pie)) 
print('pie = {0:>5.3f}   ipie = {1:>5.3f}'.format(pie, ipie)) 
print('pie = {1:>5.2f}   ipie = {0:>5.2f}'.format(pie, ipie)) 
```

### <font color='darkblue'>List comprehensions

Python does not allow operations on lists as a whole. For example, if we have two lists of numbers <tt>x, y</tt> attempt to add or divide them, or take a square of a list, as <tt>x+y</tt>, <tt>x/y</tt>, <tt>x**2</tt> will generate error. 

Python provides a compact way to generate new lists that result from mathematical operations on other lists using ***list comprehensions***, as illustrated below 

You can also read about list comprehensions in [here](https://docs.python.org/3/tutorial/datastructures.html#list-comprehensions).

```python
y = list(range(1, 11))
print(y)
# generate list containing square of the list 

x = [x**2 for x in range(1,11)]
print(x)

# compute list containing ratios x/y of x and y elements
z = [xd / yd for xd, yd in zip(x, y)]
print(z)
```

```python
# select elements of y larger than 5 and put them in a new list
g = [yd for yd in y if yd > 5]
print(g)
```

Initialization using a list comprehension:

```python
# Initialize a 2D list with nrows rows and ncols columns filled with zeroes
nrows, ncols = 5, 9
xlist = [[0]*ncols for x in range(nrows)]

# The number of rows
print(len(xlist))

# The number of columns
print(len(xlist[0]))
```

Such lists can be processed in loops like this: 

```python
for i in range(nrows):
    for j in range(ncols):
        print(xlist[i][j], end=' ') # end specifies which character to print at the end
    print('\n') # new line, so next row will be printed on a new line
```

```python
# or like this

for row in xlist:
    for col in row: 
        print(xlist[i][j], end=' ')
    print('\n')
```

List methods should be used carefully as they apply to columns by defaults. For example, applying count like this gives a seemingly strange result that there are no zeroes in the list. This is because the method scanned the rows, which contain lists of zeroes, but lists themselves are not zeroes and so are not counted. 

```python
xlist.count(0) 
```

The proper way to do this is

```python
# expression in [] creates a list of counts of zeroes in each row, which then are summed across rows
nzeroes = sum([row.count(0) for row in xlist]) 
print(nzeroes)
```

### <font color='darkblue'>Generating Gaussian-distributed pseudo-random numbers</font> 

In ex 1 of hw 2 you should have seen that sums of uniformly distributed random numbers have distribution well-approximated by the Gaussian. So you essentially already know one method of generating Gaussian random numbers. 

However, in class on Friday (see 06_pdf_gaussian_clt notebook) you saw that non-iniformly distributed random numbers can be obtained by transforming uniformly distributed random number via some mathematical operation. 

So a much more efficient way of generating Gaussian-distributed numbers is by using the <a href="https://en.wikipedia.org/wiki/Box%E2%80%93Muller_transform">Box-Mueller method</a> that works as follows. 


If $u_1$ and $u_2$ are uniformly distributed random numbers in the interval $[0,1)$, the following two numbers are drawn from the Gaussian pdf of zero mean and variance $\sigma^2=1$:


$$g_1=\sqrt{-2\ln u_1}\cos 2\pi u_2;\ \ \ g_2=\sqrt{-2\ln u_1}\sin 2\pi u_2 $$


If we want Gaussian numbers that sample Gaussian pdf with a given mean $\mu$ and standard deviation $\sigma$, we can rescale the above numbers as (where $g$ is either $g_1$ or $g_2$):

$$\tilde{g}=\mu + \sigma\,g$$


Let's check this empirically. 

**_Note_**: The method is named after George Edward Pelham Box and Mervin Edgar Muller, who <a href="https://projecteuclid.org/euclid.aoms/1177706645">proposed it in 1958</a>, although the method was previously developed by Raymond E. A. C. Paley and Norbert Wiener in 1934, in full accord with the [Stigler's law of eponymy](https://en.wikipedia.org/wiki/Stigler%27s_law_of_eponymy).

```python
from random import random 
from math import sin, cos, log, pi

def gauss_prng(mu=0, sigma=1., halfsize=1): 
    '''
    generate 2*halfsize Gaussian distributed random numbers using Box-Mueller method
    
    Returns: 
            a list of halfsize*2 floats - 
            random numbers drawn from a Gaussian pdf with mean mu and stdev sigma
    '''
    xg = []
    for i in range(halfsize): 
        u1 = random() # generate 2 random numbers
        u2 = random() # uniformly distributed in [0,1]

        # transform uniformly distributed numbers in u1 and u2 into Gaussian pdf samples
        g1 = mu + sigma*(-2.*log(u1))**0.5 * cos(2.*pi*u2)
        # g2 will stay with mu=0 and sigma=1
        g2 = mu + sigma*(-2.*log(u1))**0.5 * sin(2.*pi*u2)     
        xg.append(g1)
        xg.append(g2)
        
    return xg

# generate two vectors of nr pseudo-random numbers uniformly distributed in [0,1] (default)
nr = 5000

xg = gauss_prng(mu=0, sigma=1., halfsize=nr)
```

```python
from math import e, pi

def gaussian_pdf(x, mu=0., sigma=1.):
    """
    returns value of the Gaussian pdf with input mu, sigma for input x 

    Parameters:
    -----------
    x:     a float or a 1d numpy array of floats, input x value(s)
    mu:    a float, mean of the Gaussian (default 0)
    sigma: a float, st. deviation of the Gaussian (default 1.0)
    
    Returns:
    --------
    
    g(x): a float - value of the Gaussian pdf for input x value(s)
    """
    assert(sigma !=0.)

    return 1./(2.*pi)**0.5/sigma * e**(-0.5*((x-mu)/sigma)**2)
```

```python
def gauss_line(xmin, xmax, dx=0.01, mu=0., sigma=1.):
    '''
    construct lists of x values and corresponding Gaussian pdf values with specified mu and st dev sigma
    
    Returns:
    xp, pgauss - lists of floats containing x values in the range [xmin,xmax] and pdf values p_Gauss(x)
    '''
    pgauss = []
    xp = []
    x = xmin # start at the minimum
    dx = dx * (xmax - xmin) # increase x with step dx

    while x <= xmax: 
        pgauss.append(gaussian_pdf(x, mu=mu, sigma=sigma))
        xp.append(x)
        x = x + dx
        
    return xp, pgauss
```

```python
# construct lists contain x values and Gaussian pdf values for a grid of x 
mu, sigma = 0., 1

xp, pgauss = gauss_line(min(xg), max(xg),  mu=mu, sigma=sigma)
```

### <font color='darkblue'>Loading SDSS images from the SDSS data server

Images obtained by the SDSS are public and are arranged by tiles and frames for the Data Release 12 (DR12) in the [Flexible Image Transport System (FITS)](https://fits.gsfc.nasa.gov/) format are stored at the following URL: 
    
    https://dr12.sdss.org/sas/dr12/boss/photoObj/frames/301/
    
The URL to individual frames looks like this
    
    https://dr12.sdss.org/sas/dr12/boss/photoObj/frames/301/4263/6/frame-r-004263-6-0254.fits.bz2
    
where 4263 is the field number, 6 is the camera column (camcol) number, r is the name of the filter - a single lower-case character: u, g, r, i or z, 004263 - a 6-character string containing leading 0's and then the field number, CAMCOL number FIELD (1-digit number 1 to 6), image (frame) number 0254 consisting of 4 characters with leading 0s and number that can contain 2 to 4 digits.  
    
Functions for reading images in the FITS format exist in different Python modules, including the main astronomical Python module [Astropy](https://fits.gsfc.nasa.gov/). Example of how to use it to load images is shown below. 

**_Note_**: if you don't have [Astropy](https://www.astropy.org/) module installed, please follow instructions at its website to install it as we will be using it frequently from now on. 

```python
from astropy.io import fits

url = 'https://dr12.sdss.org/sas/dr12/boss/photoObj/frames/301/4263/6/frame-r-004263-6-0253.fits.bz2'
    
fimg = fits.open(url) # open FITS file at the URL address in the string url
image = fimg[0].data.tolist() # save image itself as a 2D list (list of lists)

# Number of rows and columns (dimensions)
nrow, ncol = len(image), len(image[0])
print('loaded FITS image of {:d} x {:d} pixels'.format(nrow,ncol))
```

If we display this image, it looks just black. 

```python
from matplotlib import pylab as plt

field, camcol, f, frame_number = '4263', '6', 'r', '0253'

# Define a single panel plot of a given size
fig, ax = plt.subplots(1,1, figsize=(10,10))
# Display image in list image using 'gray' colormap
ax.imshow(image, cmap='gray', origin='lower') 
_ = ax.axis('off') # Do not plot axis ticks
# Display plot title with the specified string
ax.set_title('SDSS: field {:s}, camcol {:s}, filter {:s}, frame {:s}'.format(field, camcol, f, frame_number))
plt.show() # Show the plot
```

We can plot the image better by setting minimum and maximum values of pixel values used in colormap to better values. For example: 

```python
fig, ax = plt.subplots(1,1, figsize=(10,10))
ax.imshow(image, cmap='gray', origin='lower', vmin=-0.13, vmax=2)
_ = ax.axis('off')
ax.set_title('SDSS: field {:s}, camcol {:s}, filter {:s}, frame {:s}'.format(field, camcol, f, frame_number))
plt.show()
```

Now sources are visible, but background is black and faint diffuse sources are difficult to see. 

This is typical when image pixel values are colored on *linear* strech of grayscale or some other [colormap](https://matplotlib.org/stable/gallery/color/colormap_reference.html), most pixels get nearly black color. For this reason such images are usually displayed on a [*logarithmic* stretch](https://matplotlib.org/stable/tutorials/colors/colormapnorms.html#sphx-glr-tutorials-colors-colormapnorms-py). 

```python
from matplotlib.colors import LogNorm

fig, ax = plt.subplots(1,1, figsize=(10,10))
ax.imshow(image, cmap='gray', origin='lower', norm=LogNorm())
_ = ax.axis('off')
ax.set_title('SDSS: field {:s}, camcol {:s}, filter {:s}, frame {:s}'.format(field, camcol, f, frame_number))
plt.show()
```

Now sources appear better, but background looks very noisy. This is because mean sky background value was subtracted from these images and some pixel values are negative. Negative values on logarithmic scale (when their logarithm is taken) produce NaN (Not a Number) values, because log of a negative value is not defined. The way to fix this is to find the minimum pixel value and subtract it from all the pixel values, so that that pixels have $\geq 0$ values. 

This can be done using loops via list elements in several different ways from the most straightforward but relatively verbose using for loops with range, to compact way using list compehension, as shown below. 

```python
imin = 1.e12 # some suitably large number for initialization

for i in range(nrow): 
    for j in range(ncol):
        if image[i][j] < imin:
            imin = image[i][j]
            
# Now subtract minimum value from each pixel
for i in range(nrow): 
    for j in range(ncol):
        image[i][j] -= imin
```

or this can be done as 

```python
imin = 1e12

for row in image: 
    for x in row: 
        if x < imin: imin = x

flat_image = []
for i, row in enumerate(image): 
    for j, x in enumerate(row): 
        image[i][j] -= imin 
        flat_image.append(image[i][j])
```

or this can be done as follows using [list comprehension](https://www.w3schools.com/python/python_lists_comprehension.asp)

```python
# Determine the minimum of image pixel values using "list comprehension"
imin = min([min(row) for row in image]) # min() is a built-in function for finding minimum of a simple list

# Subtract the minimum from each image pixel using nested list comprehension
image = [[(x - imin) for x in row] for row in image]
```

Displaying image now makes it look much nicer with many more faint sources visible. 

```python

fig, ax = plt.subplots(1,1, figsize=(10,10))
ax.imshow(image, cmap='gray', origin='lower', norm=LogNorm())
_ = ax.axis('off')
ax.set_title('SDSS: field {:s}, camcol {:s}, filter {:s}, frame {:s}'.format(field, camcol, f, frame_number))
plt.show()
```

We can also use "inverted" gray scale with colormap "Greys", in which brighter regions are displayed as darker gray. 

```python
fig, ax = plt.subplots(1,1, figsize=(10,10))
ax.imshow(image, cmap='Greys', origin='lower', norm=LogNorm())
_ = ax.axis('off')
ax.set_title('SDSS: field {:s}, camcol {:s}, filter {:s}, frame {:s}'.format(field, camcol, f, frame_number))
plt.show()
```
