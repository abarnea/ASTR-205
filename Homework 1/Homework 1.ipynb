```python
def myfactorial(n):
  if n<0:
    print("Invalid input! Please rerun the program and enter a non-negative integer.")
  elif n==0:
    return 1
  else:
    for i in range(1,n):
      if i<n:
        n *= i
    return n

myfactorial(int(input("Pick a number: ")))
```

In the worst case scenario, the fraction of photometric nights is $p=0.4$, so if we treat this as our probability p and utilize the binomial probability function from Exercise 2, we can calculate the probability we will get four photometric nights out of our seven:

```python
print("The probability of getting the worst case scenario of photometric nights is", pbinomial(0.4,7,4))
```

As such, in the worst case scenario, the probability that our observational campaign at Las Campanas Observatory succeeds at getting four photometric nights out of seven and not having to reapply is not terrible at $\boxed{P\approx 0.19}$ . We can use the Full Width at Half Maximum (FWHM) to find out how narrow our binomial distribution is -- i.e. quantify our spread of data points and see how realistic/unrealistic they are. Tests for COVID-19 (or any other disease) are characterized by their probability to produce positive result when disease is present, $P(+\vert C)$ (where $C$ stands for COVID presence),  which is called *test sensitivity*. 

Another important characteristic of a test is *specificity* defined as the probability that a test result will be negative when disease is not present, $P(-\vert N)$ (where $N$ stands for "No disease present"). Specificity is a complement probability of the probability to get positive test result when disease is not present, $P(+\vert N)=1-P(-\vert N)$, which is called the *false positive rate*.

Probability for a random member of a community to have COVID, P(C), can be conservatively approximated by the current positivity rate (the fraction of people in the community who test positive). 

**Task.** 
Suppose you know when you are taking a COVID test that its sensitivity is $P(+\vert C)=0.97$, specificity is $P(-\vert N)=0.995$ and the current COVID positivity rate on campus is 1\% so $P(C)= 0.01$. 

If you test positive with this test, compute the probability that you *actually* have COVID-19? 

How does the result change if the test specificity is $P(-\vert N)=0.97$ instead? What can you conclude about importance of knowing specificity of such tests?

*Hint:* use the Bayes theorem and the "law of total probability", which were discussed on Monday and are described in [02_probability](https://drive.google.com/file/d/18r-zCP4W67VW5HAOElvmpXkTB2Q_kX88/view?usp=sharing) notebook. 

*Note:* that this exercise does not require much coding, but it requires a derivation of equation for the corresponding probability and calculation of that probability. You should present your argument, equations, and calculations of the probabilities.

In order to find the probability that you have COVID given that you tested positive $P(C\vert +)$, it will suffice to utilize Bayes' Theorem to do so. From Bayes' Theorem, we have that:

$$P(C\vert+)=P(+\vert C)\cdot\frac{P(C)}{P(+)}$$

We are given both the probability of having COVID on campus, $P(C)=0.01$ and the probability of testing positive given that you have COVID, $P(+\vert C)=0.97$, so all we need to find is the probability of testing positive $P(+)$. There are two scenarios where you can test positive -- the first is if you have COVID and test positive $P(+\vert C)$, and the second is that you don't have COVID and test positive $P(+\vert N)$. Using a simple probability tree, we find that: 

$$P(+\text{ and }C)=P(+\vert C)\cdot P(C)\\ P(+\text{ and }N)=P(+\vert N)\cdot P(N)$$

Since the probability of testing positive $P(+)$ is the sum total of probabilities of testing positive given you have COVID and not having COVID (as those are the only two scenarios), then we can simply add the two probabilities $P(+\vert C)$ and $P(+\vert N)$ to get $P(+)$, which gives us: 

$$P(+)=P(+\text{ and }C)+P(+\text{ and } N)=P(+\vert C)\cdot P(C)+P(+\vert N)\cdot P(N)$$

Plugging this back into our equation for $P(C\vert +)$ from earlier to find the probability of having COVID given that you tested positive, we find:

$$P(C\vert +)=P(+\vert C)\cdot\frac{P(C)}{P(+\vert C)\cdot P(C)+P(+\vert N)\cdot P(N)}\\=\frac{0.97\cdot 0.01}{0.97\cdot 0.01+0.005\cdot 0.99}=0.6610\approx \boxed{0.66}$$

If the test specitivity is $P(-\vert N)=0.97$ instead of $P(-\vert N)=0.995$, then all we need to do is change that value in our equation for $P(C\vert +)$. Doing so, we find a new probability of having COVID given that you tested positive of:

$$P(C\vert +)=\frac{0.97\cdot 0.01}{0.97\cdot 0.01+0.03\cdot 0.99}=0.2462\approx \boxed{0.25}$$
