### <font color='blue'>Loops:</font> 

**Background info:** When we know how many times we want to repeat a given calculation/code, we can use a for loop that is usually written in this form: 

    for i in range(nbegin,nend,nstep): 
        # some code here
        
where <tt>i</tt> is called the iteration variable, <tt>nbegin</tt> and <tt>nend</tt> are the limits of the range of index i that we want to go through during loop cycles (called "iterations"). Note that the index will go through <tt>nbegin</tt> to <tt>nend-1</tt>. 

The function <tt>range</tt> returns a sequence of numbers from <tt>nbegin</tt> to <tt>nend-1</tt> with increment <tt>nstep</tt>

For example, for <tt>i</tt> to cycle <tt>n</tt> times from 1 to <tt>n</tt> we do (note again that to go to <tt>n</tt> we need to have the upper limit to be <tt>n+1</tt>). 


```python
n = 10
for i in range(1,n+1):
    print(i)
```

```python
for i in range(1,n+1,2):
    print(i)
```

### <font color='blue'>Simple functions:</font> 

Code can be wrapped into functions for convenience to allow for re-using it with different parameter values. 
For example, a function <tt>func</tt> with parameter <tt>x</tt> can be defined as

    def func(x):
        #do something with x
        return result
        
Last line could return a single variable: <tt>return result</tt>
Several of them: <tt>return result1, result2, result3</tt>
Or nothing at all: <tt>return</tt>

In the latter case the function returns a special value <tt>None</tt>. 

Function below for example simply prints its argument:

```python
def print_arg(x):
    print(x)
    return 

print_arg(3)
print_arg('Bayes theorem: P(B|A) = P(A|B)*P(B)/P(A)')
print_arg('Total probability law: P(A) = P(A|B)*P(B) + P(A|not B)P(not B)')
```
