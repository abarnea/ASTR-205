## <font color='darkblue'>Probability as a measure of plausibility or "degree of belief"
    
Development of probability theory in the 17th-19th centuries was driven by applied questions related to specific questions related to "games of chance" (dice, cards, etc.). 
    
In the 20th century a more nuanced view of probability was developed as part more rigorous formulations of probability theory. 
    
The theory is based on the view expressed in the title of this slide and several basic axioms, from which everything else follows. 


Several such theories were developed in the 20th century with subtle but important differences in the definition/interpretation of what probability means. There were intense debates about which interpretation of probability is preferred, in particular between [H. Jeffreys](https://en.wikipedia.org/wiki/Harold_Jeffreys) and [E.T. Jaynes](https://en.wikipedia.org/wiki/Edwin_Thompson_Jaynes). None of this really matters for this course or for applications of statistics in practice, but if you like to deep dive into foundations of a subject, you can find detailed account of these different interpretations [here](https://plato.stanford.edu/entries/probability-interpret/). 


We will follow the first and most commonly used set of axioms developed by Andrey Kolmogorov. 

### <font color='darkblue'>Axioms of probability

Mathematician/physicist [Andrey Kolmogorov](https://en.wikipedia.org/wiki/Andrey_Kolmogorov) in his book "*Foundations of the Theory of Probability*" published in 1933 showed that with a set of 3 assumptions (*axioms*) one can construct a self-consistent and useful probability theory.

### <font color='darkblue'>Axioms of probability


**_Axiom 1._** Probability of an event (or outcome or proposition) is a non-negative real (floating point) number $P(E)>=0$.


We want to have a measure that spans the range between the logical 0 and 1 and so probability must be within the interval $[0,1]$ *inclusive.*

**_Axiom 2._** Probability of at least one outcome in the entire possible range of outcomes is 1.


For example, we can always have outcomes A and "not A" and probability of either occurring must be 1 according to this axiom. This axiom stems from the fact that certain events are subject to logic and the fact that one of all possible events must occur is True, which corresponds to 1 in logical variable. 

**_Axiom 3._** For *mutually exclusive* events $E_1$, $E_2$, $\ldots$, $E_N$  probability of any of these events occurring is 

$$P({\rm any\ of}\ E_i) = P(E_1+E_2+\ldots +E_N) = P(E_1)+P(E_2)+\ldots + P(E_N) = \sum\limits_{i=1}^N P(E_i).$$



Recall that in mathematical logic "+" stands for "or". 

For example, when we throw a coin, heads and tails are the only outcome (or if you want to be exhaustive, heads, tails, standing on the edge). 

For dice, the entire range of outcomes is getting 1, or 2, or 3, or 4, or 5, or 6 (note that these outcomes are *mutually exclusive* in the sense that if one occurs others cannot occur at the same time). 

From these axioms, various rules of probability can be derived. 

* From Axioms 2 and 3: $P(\bar{E}) = 1 - P(E)$, where $\bar{E}$ is "not $E$". 


* From Axiom 1, $P(\bar{E})\geq 0$ and from $P(\bar{E}) = 1 - P(E)\geq 0$, follows that $P(E)\leq 1$  

Using these we can go on and derive other expressions.  

### <font color='darkblue'>Probability of either of two non-exclusive events $A$ or $B$ occurring
    
Because events are not exclusive, we can have $P(A)+P(B)>1$. So **_summing probabilities of non-exclusive events will give an incorrect answer!_** 
    
Instead:   

$$P(A + B) = P(A) + P(B) - P(A\cdot B),$$
    
where $ P(A\cdot B)$ is the probability of $A$ and $B$ occuring. 

**Example:** probability that it will not rain in Chicago at the end of September and probability that day-time high termperature will be above 60F are both $>0.5$ so their sum would be $>1$. However, a day without rain does not exclude the possibility that it will also be warm. So these are non-exclusive events and probability of either one occuring needs to take into account probability that they both occur.  

This can be understood using [Venn diagram](https://en.wikipedia.org/wiki/Venn_diagram), which shows probabilities of $A$ and $B$ as overlapping regions with size of region area reflecting the magnitude of each probability. Size of region of overlap reflects $P(A\cdot B)$. 

It's possible to extend this approach to more than two events, but the regions of overlaps start to multiply quickly and one has to be careful to account for all of them carefully

The [general rule](https://www.thoughtco.com/probability-union-of-three-sets-more-3126263) is  

* Add the probabilities of the individual events.

* Subtract the probabilities of the intersections of every pair of events.

* Add the probabilities of the intersection of every set of three events.

* Subtract the probabilities of the intersection of every set of four events.

Continue this process until the last probability is the probability of the intersection of the total number of sets that we started with


### <font color='darkblue'>Probability of both non-exclusive events $A$ *and* $B$ occurring

Given that events are not exclusive, in general we can $A$ and $B$ both occuring. These events may either be independent or may depend (or "conditioned") on each other. 
    
**_Conditional probability_** of $A$ occuring given $B$ occurs (and thus $P(B)>0$) is usually denoted as $P(A|B)$ is given by 
    
$$P(A\vert B) = \frac{P(A\cdot B)}{P(B)}$$ 

This expression for $P(A|B)$ can be derived by considering the following situation. Suppose we have two adjacent pieces of land owned by two different families $A$ and $B$. They also have an overlap area the ownership for which is disputed. 

If they find a treasure on the piece of land claimed by $B$ as its own, what is the probability that the treasure is found within the disputed area? This probability is $P(A\vert B)$ and is given by the ratio of areas of disputed area and the total area claimed by family $B$, which is what equation above is. 



### <font color='darkblue'>Probability of both non-exclusive events $A$ *and* $B$ occurring

and thus probability of $A$ *and* $B$ to occur is 
    
$$P(A\cdot B) = P(A\vert B)\,P(B)$$

or via symmetric argument: 

$$P(A\cdot B) = P(B\vert A)\,P(A)$$


If $A$ and $B$ are independent, we can just drop $\vert$ and what's to the right of it: 

$$P(A\cdot B) = P(A)\,P(B)$$

probability two independent outcomes both occurring is equal to the product of their individual probabilities. 

**Example:** probability of getting tails on a coin toss is $0.5$, the probability of getting two tails in a row is $0.5\times 0.5=0.25$. 

**Example:** probability of getting two 6s on two rolls of dice is $1/6\times 1/6=1/36\approx 0.0278$ 

For independent outcomes this extends for an arbitrary number of different outcomes, i.e. 

$$P(A\cdot B\cdot C) = P(A)\,P(B)\, P(C)$$

etc. 

**_Note_**: Incidentally, from the above equivalent expressions for $P(A\cdot B)$ we have

$$P(A\cdot B) = P(A\vert B)\,P(B)= P(B\vert A)\,P(A), $$

which gives the famous **_Bayes theorem_** (we will discuss it later in a different context):

### <font color='darkblue'>Probability of $n$ outcomes in $N$ experiments 

Suppose we want to ask a question of what is probability of $n$ outcomes in $N$ trials? 
    
For example, what is the probability that there will be rain during 4 days out of the next seven? (assuming that probability of rain each day is independent) 

We are now equipped to do this. 

If the probability to rain on a given day is $p$, then probability to rain during 4 specific days **_and_** not rain during the other 3 is $p^4(1-p)^3 = p^4(1-p)^{7-4}$. 

We have many different specific combinations of rainy and not rainy days and the question did not specify which days it should rain, so we are considering probability of *any* of these possibilities and thus need to sum up their probabilities (Probability Axiom 3). 

In this case there are 

$$\binom{7}{4} = \frac{7!}{4!(7-4)!}$$

combinations of four rainy days out of 7 (see Section 1.3.2 of [David Morin's book draft](http://www.people.fas.harvard.edu/~djmorin/probability.pdf)). 

```python
p = 0.26

7*6*5*4*3*2*1 / (4*3*2*1) / (3*2*1) * p**4 * (1-p)**3
```

### <font color='darkblue'>The *binomial distribution*

In general for $n$ outcomes out of $N$ trials

$$\binom{N}{n} = \frac{N!}{n!\,(N-n)!}$$

and probability of having $n$ outcomes with probability of individual outcome of $p$ in $N$ trial is 

$$p_N(n) = \frac{N!}{n!\,(N-n)!}\, p^n\,(1-p)^{N-n},$$

which is called the **_binomial distribution_**. 

**_Note:_** This distribution was first derived by Jacob Bernoulli in his 1713 book ["*Ars Conjectandi*"](https://en.wikipedia.org/wiki/Ars_Conjectandi)

**_Note:_** The binomial distribution name is because $\binom{N}{n}$ is called  the *binomial coefficient*. This coefficient arises in the *binomial expansion* (*bi* since there are two variables $a$ and $b$ in the expansion):

$$(a+b)^N = \binom{N}{0}a^N + \binom{N}{1}a^{N-1}b+\ldots +  \binom{N}{N-1}ab^{N-1} + \binom{N}{N}b^N = \sum\limits_{n=0}^N\binom{N}{n}\,a^{N-n}b^n.$$
