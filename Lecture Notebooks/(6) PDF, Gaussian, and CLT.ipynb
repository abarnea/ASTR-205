Probability is ***dimensionless***. This means that $dx$ must be multiplied by something that has units that are inverse of $x$, so that the product (probability) is dimensionless. 

Given that probability for a uniform distribution in $[0,1]$ to be within $[0,0.1]$ is 0.1, then we are multiplying $dx$ by a function, $p(x)=1$, but which must have dimensions of inverse $x$. That is, if we are drawing random masses in kg, then $p(x) = 1\,\rm kg^{-1}$. 

$p(x)$ is called the probability density - it measures the amount of probability per unit $dx$, so that $p(x)dx$ gives the probability for $x$ to be in the interval $[x, x+dx]$.

### <font color='darkblue'>Probability density function (pdf)
    
for a continuous random variable $x$ is $p(x)$, such that *probability* for $x$ to be in the interval $[x, x+dx]$ (where $dx$ is some very small interval of $x$) is equal to $p(x)dx$.

### <font color='darkblue'>pdf normalization


pdf needs to conform to the axioms of probability, from which it follows that the total probability over the entire possible range of $x$ is 1.

$$\int\limits_{-\infty}^\infty p(x) dx = 1,$$

where integral is taken over all possible $x$ values. 

### <font color='darkblue'>Definition: *statistic*</font>

A *statistic* is a number computed for a pdf or for a sample of random numbers drawn from a pdf. 

The most commonly used statistics are the mean, variance, and standard deviations, but there are other statistics that are useful in particular situations. 

### <font color='darkblue'>Definitions of the mean and variance of $x$ drawn from pdf $p(x)$</font>

Expectation value: $E(x) \equiv \mu = \int\limits_{-\infty}^{\infty} x p(x) dx$ 

central variance: $\sigma^2\equiv \mu_2 = \int\limits_{-\infty}^{\infty} (x -\mu)^2 p(x) dx$

### <font color='darkblue'>higher central moments of $x$ drawn from pdf $p(x)$</font>


$\mu_3 = \int\limits_{-\infty}^{\infty} (x -\mu)^3 p(x) dx$ and associated quantity called *skewness*: $\tilde{\mu}_3(x) = \mu_3/\sigma^3$.

$\mu_4 = \int\limits_{-\infty}^{\infty} (x -\mu)^4 p(x) dx$ and associated quantity called *kurtosis*: $\tilde{\mu}_4(x) = \mu_4/\sigma^4$.

etc.

### <font color='darkblue'>Uniform pdf</font>

Uniform pdf defined on the interval $[a,b]$ is 

$$p(x) = \begin{cases} 1/(b-a),\ \mathrm{for}\ x\in[a,b]\\
0,\ \mathrm{otherwise}\end{cases}$$

### <font color='darkblue'>Uniform pdf in $[a,b]$</font>

$$p(x) = \begin{cases} 1/(b-a),\ \mathrm{for}\ x\in[a,b]\\
0,\ \mathrm{otherwise}\end{cases}$$

mean: $\mu = \int\limits_{-\infty}^{\infty} x p(x) dx = \frac{1}{b-a}\,\int\limits_0^1 x dx=\frac{1}{b-a}\frac{x^2}{2}\vert^b_a=\frac{1}{2}\,(b + a)$ 

central variance: $\mu_2 = \int\limits_{-\infty}^{\infty} (x -\mu)^2 p(x) dx = \frac{1}{b-a}\,\int\limits_a^b \left(x-\mu\right)^2dx = \frac{1}{b-a}\,\int\limits_{a-\mu}^{b-\mu}\tilde{x}^2d\tilde{x}= \frac{1}{b-a}\,\frac{\tilde{x}^3}{3}\vert^{b-\mu}_{a-\mu}=\frac{1}{3(b-a)}\,\left[(b-\mu)^3 - (a-\mu)^3\right]$


### <font color='darkblue'> Uniform distribution in the interval $[0,1]$</font>

* a random number drawn from such distribution has probability of 1 to be within the above interval and probability 0 to be outside of it. 

* If we take a sub-interval, such as $[0, 0.1]$, the probability for a given number to be within it is 0.1.
In general, probability for sub-interval $[x,x+dx]$ that is fully enclosed in $[0,1]$ is $dx$.  

But what if $x$ is a physical variable with units (dimension) associated with it? For example, suppose I draw random numbers that are temperatures. Does probability then have units of temperature? 

### <font color='darkblue'>Uniform pdf in $[0,1]$</font>

$$p(x) = \begin{cases} 1,\ \mathrm{for}\ x\in[0,1]\\
0,\ \mathrm{otherwise}\end{cases}$$

mean: $\mu = \int\limits_{-\infty}^{\infty} x p(x) dx = \int\limits_0^1 x dx=\frac{x^2}{2}\vert^1_0=\frac{1}{2}$ 

central variance: $\mu_2 = \int\limits_{-\infty}^{\infty} (x -\bar{x})^2 p(x) dx = \int\limits_0^1 \left(x-\frac{1}{2}\right)^2dx = \int\limits_{-1/2}^{1/2}\tilde{x}^2d\tilde{x}= \frac{\tilde{x}^3}{3}\vert^{+1/2}_{-1/2}=\frac{1}{24}+\frac{1}{24}=\frac{1}{12}$

$\mu_3=0$ by symmetry

```python
from matplotlib import pylab as plt

def plot_hist(x, bins = 'auto', figsize = (6,6), density = True, xlabel = 'x'):
    '''
    Plot histogram of values in x 
    
    where x is either a list or numpy array
    the other argument are simply passed to Pylab function hist 
    '''
    plt.figure(figsize = figsize) # define figure and its dimensions
    plt.xlabel(xlabel) # x-axis label
    plt.ylabel('histogram') # y-axis label
    plt.hist(x, density = density, color = 'slateblue', bins = bins) # produce histogram of xs values
    plt.show() # show the entire plot
```

```python
from random import random 
def random_uniform(low=0., high=1., size=1):
    '''
    produce size pseudo-random numbers uniformly distributed in the interval [low, high]
    
    returns the list of the numbers of length size
    '''
    assert(size >= 1) # make sure that size is >=1
    assert(high > low)
    
    x = [None] * size
    ba = high - low
    for i in range(size): 
        x[i] = ba * random() + low 
    return x
```

```python
from random import random 

def uniform_mean_mc(ns, nr): 
    '''
    computes ns sums of nr uniformly distributed random numbers
    
    Parameters: 
        ns - int, number of sums to compute
        nr - int, number of random numbers to use in the sums
    Returns:
        xs - list of length ns containing sums of nr random numbers
    '''
    xs = [None] * ns

    for n in range(ns): # loop to obtain ns sums
        xsum = 0.
        for i in range(nr): # loop to obtain current sum
            xd = random() - 0.5
            xsum += xd
        xs[n] = xsum # record the sum in xs list
    return xs
```

```python
a, b = 1, 3 # so we expect p(x) = 1/(b-a) within interval [a,b]
nr = 100000
# produce a list x of nr uniformly distributed random numbers using Python's random.random() function
xs = random_uniform(low=a, high=b, size = nr)

for i, xd in enumerate(xs):
    xs[i] = xd**2
plot_hist(xs, density=True, bins='auto')
```

#### <font color='darkblue'>The Central Limit Theorem:
    
If we get a certain number $N_r$ of random numbers from  a random process, then the probability distribution for the sum of their values will be approximately described by the Gaussian (aka “normal”) distribution. As $N_r$ increases, the distribution of the sums will be described by the Gaussian pdf better and better.


Gaussian pdf with mean $\mu$ and variance $\sigma^2>0$ is given by the following expression: 

$$p(x) = \frac{1}{\sqrt{2\pi\sigma^2}}\, \exp\left[-\frac{(x-\mu)^2}{2\sigma^2}\right].$$ 

**_Note_**: there are certain random distributions for which the above is not true, but these are exceptions. For most random processes the above theorem will apply. 

***Note.*** Because $x-\mu$ is squared in the exponent, the distribution is symmetric for $x<\mu$ and $x>\mu$. It peaks at $x=\mu$ and falls off rapidly on both sides of the peak.



The distribution is named after Carl Fridrich Gauss - a German astronomer and mathematician who (re-)discovered it in the early 1800s, although just like for the Poisson distribution the first person who discovered it was actually another French mathematician, [Abraham de Moivre](https://en.wikipedia.org/wiki/Abraham_de_Moivre).

```python
from matplotlib import pylab as plt
import numpy as np 

def plot_gaussian(xmin, xmax, mu = 0, s2 = 1., nplot = 1000): 
        xe = np.linspace(xmin, xmax, nplot) # use numpy functions to generate a vector of evenly spaced x values
        ge = np.exp(-0.5 * (xe - mu)**2 / s2) / np.sqrt(2. * np.pi * s2) # and compute Gaussian pdf for them
        plt.plot(xe, ge, label='Gaussian pdf', color='purple', lw=2.)
        plt.legend(loc='upper left', fontsize = 12) # place legend with labels that were defined

plt.figure(figsize = (8,6))
plt.xlabel('x', fontsize = 15)
plot_gaussian(-5, 5)
plt.show()
```
