### <font color='darkblue'>Consider the question of estimating interval between shuttle arrivals</font>


*Suppose you arrive at a University shuttle stop during regular shuttle operating hours and a shuttle arrives after you wait for 10 minutes. Do you think it is possible to estimate the mean interval between shuttle arrivals from this single measurement?*


If you waited 10 minutes for a bus, you can immediately conclude that $\tau> 10$ minutes. However, we can do better (ostensibly) than this using statistical methods. 

If you assume that buses arrive at exact time intervals $\tau$ and there was nothing special about your arrival time (for example, we did not see that previous bus was leaving as we were approaching the stop). This assumption means that the distribution of time $t$ between the last bus and your arrival is a uniform random number drawn from a uniform pdf: 

$$p(t\vert \tau)dt = \begin{cases}\frac{1}{\tau},\ {\rm for}\ 0<t\leq\tau \\
0,\ \rm otherwise\end{cases}$$

where as before $1/\tau$ normalization ensures that $\int_{-\infty}^\infty p(t\vert\tau)dt =1$, as required for a pdf. 

**_Note_**: the assumptions that allow us to derive a given pdf constitute *the model* that we are using to make *statistical inference* about a process. Model always has some assumptions and *parameters.* In this case the parameter is the interval between bus arrivals $\tau$. 

### <font color='darkblue'>Likelihood pdf 
    
The pdf that involves experimental measurements and model parameters is called the *likelihood pdf* in this context. In this example, $p(t\vert\tau)$ is the likelihood pdf for the waiting time $t$ in the model where buses arrive at the exactly evenly spaced time interval $\tau$. 

The measurements are result of the experiment and do not depend on any model, but the likelihood pdf of the measurement does depend on the model we adopt (for example, the simplistic model above will be modified to allow Gaussian scatter in bus arrival times, which will modify the likelihood pdf). 
    
To explicitly indicate this dependence, the likelihood pdf is usually expressed as a *conditional* pdf depending on the model assumptions and parameters: in this case this can be expressed as $p(t\vert \tau)$ with assumptions about strict intervals implied and manifested in the functional form of the pdf. 
    
As with conditional probabilities, the quantities to the right of $\vert$ are simply indicating what the pdf depends on (or *conditioned on* in the stat jargon). The actual variable for which the pdf is defined is to the left of $\vert$ (in this case it is experimental measurement of the time you had to wait for the bus, $t$). 

### <font color='darkblue'>Model parameter inference using the maximum likelihood method


The likelihood pdf often has a clear peak (e.g., when it is described by the Gaussian, Student $t$, Poisson, etc.) and *the maximum likelihood method* derives conclusions about likely model parameter values by using the $\alpha$ *confidence interval*, where $\alpha$ is the fraction of the total probability (total area under the pdf curve) that is enclosed within the interval centered on the pdf maximum. $\alpha$ is often expressed as percentage (which is $100\alpha\%$): e.g., "50% confidence interval", "95% confidence interval", etc. 

Operationally, one finds the maximum of the likelihood pdf and then determines how far away likelihood argument has to be to the left of the maximum to enclose $\alpha/2$ of the total probability of 1, and how far to the right of the maximum the argument needs to be to enclose the other $\alpha/2$. 

In the illustration below, $\alpha=0.95$ and areas in the left and right tails of the distribution are $(1-\alpha)/2=0.025$.

### <font color='darkblue'>The maximum likelihood estimate of the bus arrival interval
    
The uniform pdf defined in this problem does not have a distinct peak, but it has a well defined range of $t$ where it is not zero and we can define the confidence interval around the middle of the interval $t_{\rm mid}=\tau/2$. 

Cumulative distribution function of the uniform pdf above gives the probability that you will have to wait less than a given time $t_{\rm wait}$: 

$$P(<t_{\rm wait})=\int\limits_0^{t_{\rm wait}}{dt\over \tau}=\begin{cases}{t_{\rm wait}\over\tau},\ {\rm for}\ 0<t_{\rm wait}\leq\tau \\
1,\ \rm t_{\rm wait}>\tau\end{cases}$$

If you assume that your wait time $t_{\rm wait}$ falls within the *central* 50\% of the probability distribution, the 50% confidence interval is 

$${t_{\rm wait}\over\tau}\in [0.25, 0.75]$$

Thus, for the observation of $t_{\rm wait}=10$ min, 

$${10\,{\rm min}\over\tau}\in [0.25, 0.95]$$

the corresponding confidence interval for $\tau$ is then 


$$\tau_{50}=10\, {\rm min}\times\left[{1\over0.75}, {1\over0.25}\right]=10\, {\rm min}\times \left[1.333, 4\right]=[13.33, 40]\,\rm min$$


That is this observations implies that $50\%$ confidence range for the interval between buses is between 13.3 and 40 minutes. 

*Does this seem like a sensible estimate?*

### <font color='darkblue'>Common sense expectation for the bus waiting time</font>


You've probably waited for buses/shuttles in your life before. 
Any time you arrive at a shuttle stop and decide to wait for a shuttle, you've already made an estimate
for an average time between shuttle arrivals even though you may not be
aware of it. 

If this was not the case, then your wait time could be
arbitrary - say, 5 hours. So decision to wait indicates that your
estimate excludes arrival intervals more than, say, $\sim 1$ hour (or whatever your
 tolerance threshold is).
 
 So when you have a measurement that shuttle
arrived after 10 min, you can definitely make an estimate for the confidence interval of $\tau$, because you are able to make one even without it. 

The measurement can be used to improve whatever your estimate based on *prior* experience. However, you should be very suspicious if the measurement ***contradicts*** it. Either you've discovered a life-changing new behavior of buses that will have dire consequences for your transportation options, or (more likely) the statistical estimate made is flawed.  

### <font color='darkblue'>Maximum likelihood versus common sense</font>

The estimate that was done is using what is known as the *maximum likelihood method*. We make some sensible assumptions about the process, which allow us to adopt a pdf (called *likelihood pdf* in this context) for the parameter of interest (uniform pdf in the case of the bus wait time question). And then we use this pdf to estimate the parameter value where the likelihood pdf has maximum and the confidence interval enclosing some fraction of the total probability around the maximum (or the middle of the interval where pdf is not zero like we did, because uniform pdf does not have a distinct maximum). 

How is it that our common sense gives us a better estimate of the confidence limits than the formal maximum likelihood method? 


### <font color='darkblue'>Mathematical version of what our brain does: prior pdf</font>

We use our ***prior*** experience and information about the event. This is used to form an expectation for the probability distribution function of the shuttle arrival, even though we do not formulate it in a rigorous mathematical fashion in our daily life. 

We use this approach in almost everything we do. Sometimes this is referred to as a "gut feeling" or "intuition". 


Our brain estimates probability that the interval between shuttles, $\tau$, using some sort of pdf based on the current conditions (e.g. night or day) and previous experience or information about shuttles that we know. 

Mathematically, we have to adopt some form for it and in statistics this is called the ***prior pdf***. 

Buses don't arrive at exact intervals (unless you are in Switzerland). 

To make the model more realistic, let's assume that our expectation is that the actual interval between shuttle arrivals, $\tau$, is described by the Gaussian pdf with the mean $\bar{\tau}$ and st. deviation of $\sigma_\tau$: $p_{\rm G}(\tau\vert \bar{\tau}, \sigma_\tau)$. 


So then, when we are *heading to the shuttle stop*, we expect that with 96% confidence level our wait time will be between $\bar{\tau}-2\sigma_\tau$ and $\bar{\tau}+2\sigma_\tau$. 

When we arrive at a shuttle stop, if our arrival time is not special (we did not see shuttle leaving and we did not see the shuttle when we approach the sotp), the *likelihood* that we wait time $t$ until the next shuttle is $t/\tau$, as before, but given that our expectation that $\tau$ is a time drawn from  $p_{\rm G}(\tau\vert\bar{\tau},\sigma_\tau)$, the actual pdf of the wait time is: 

$$\frac{t}{\tau}\,p_{\rm G}(\tau\vert \bar{\tau},\sigma_\tau)$$

Because the likelihood is multiplied by the Gaussian prior pdf, $p_{\rm G}(\tau\vert \bar{\tau},\sigma_\tau)$, which decreases rapidly for $\tau\gg \bar{\tau}$, our estimate of the confidence regions for the estimate of $\tau$ given the $t_{\rm wait}$ we actually needed to wait to a much narrower range. 

### <font color='darkblue'>Credible regions vs Confindence intervals</font>

In the Bayesian analysis the analogs of the confidence intervals are called *credible regions*. 

This is not to create yet more confusing stat jargon, but to distinguish the interval obtained from the posterior pdf, from the confidence intervals obtained from the likelihood pdf in the maximum likelihood method. 


### <font color='darkblue'>Minimizing $-\ln p$ instead of maximizing $p$</font>

Logarithm of the pdf is usually used in finding the paramater values that maximize the pdf, because pdf can be easily very small during the search for parameters that minimize it and to avoid *floating point underflow* due to limited number of bits in the floating numbers, we work with the natural log instead. 

Also, Python module SciPy provides *minimization* functions, which find for which values of function arguments the value returned by the function is *minimized.* However, if we want to find the maximum of a function, this is equivalent to finding the minimum of its negative. So we can use the the minimization function [<tt>scipy.optimize.minimize</tt>](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize.html) to minimize 

$$-\ln\left[\frac{t}{\tau}\,p_{\rm G}(\tau\vert \bar{\tau},\sigma_\tau)\right]=-\ln\frac{t}{\tau} -\ln p_{\rm G}(\tau\vert \bar{\tau},\sigma_\tau).$$

The latter transformation uses the fact that log of a product is sum of the logarithms: $\ln[p(x)p(y)] = \ln p(x) + \ln p(y)$. 


### <font color='darkblue'>Putting this all together to get credible regions</font>


First the function for the product of the likelihood pdf and prior pdf. This function computes not the posterior pdf but negative natural log of the posterior pdf. 


```python
def loglike_prior(tau, twait, tau_mean, sigma_tau):
    '''
    -ln of the product of the likelihood pdf and prior pdf 
    for the bus shuttle interval model inferrence
    '''
    assert(twait>0)
    assert(tau_mean>0); assert(sigma_tau>0)
    
    # likelihood pdf:
    pwait = twait / np.abs(tau)
    # -ln(likelihood pdf * prior pdf)
    stau2 = sigma_tau**2
    return (-np.log(pwait) + 
            0.5 * (np.log(2*np.pi) + np.log(stau2) + 
                   ((tau-tau_mean)/sigma_tau)**2))
```

Now find the value of $\tau$ at which posterior pdf has maximum. 

```python
from scipy.optimize import minimize

p0 = [10] # initial value for tau during minimization

'''
    twait is our observation of the bus arrival
    tau_mean, sigma_tau our general expectation for bus interval
    and its uncertainty
'''

twait, tau_mean, sigma_tau = 10, 15, 5

'''
    scipy.optimize.minimize fins values of arguments of loglike_prior
    at which the function has minimum; p0 is initial tau value
    method is minimization algorithm to use
    tol is accuracy with which minimum will be identified
'''
res = minimize(loglike_prior, p0, args=(twait, tau_mean, sigma_tau), 
               method='nelder-mead', tol=1.e-10)
tau_max = res.x[0]
print('estimate of the interval between shuttles = {:>.4g} minutes'.format(tau_max))
```

```python
def get_xpfrac(x, pdf, pfrac):
    '''
    Estimate the value of pdf argument below which pfrac of the total
    probability lies. 
    '''
    pcum = np.cumsum(pdf) / np.sum(pdf)
    inds = (pcum < pfrac)
    xd = x[inds]
    return xd[-1]
```

Now estimate the 68.27% and 95% credible regions 

```python
taumin, taumax, ntau = 0.1, 50., 500
taug = np.linspace(taumin, taumax, ntau)

# value of the pdf at the maximum
ptau = np.exp(-loglike_prior(taug, twait, tau_mean, sigma_tau))

# 95% credible region
p025, p975 = (get_xpfrac(taug, ptau, 0.025), 
              get_xpfrac(taug, ptau, 0.975))

# 68.27% credible region
p341, p657 = (get_xpfrac(taug, ptau, 0.1573), 
              get_xpfrac(taug, ptau, 0.8427))

print('max. likelihood at tau = {:.4g} minutes'.format(tau_max))
print('68.27% confidence limits: {:.4g} - {:.4g} minutes'.format(p341, 
                                                                 p657))
print('95% confidence limits: {:.4g} - {:.4g} minutes'.format(p025, 
                                                              p975))
```

Now let's plot posteriod pdf with these credible regions shown as shaded regions. 

```python
def plot_posterior(xg, pdfg, figsize=(10,5), xlabel='x'):
    ''' plot posteriod pdf for gridded values of paramer xg and its pdf in pdfg '''

    imax = np.where(pdfg == pdfg.max())
    xmax = xg[imax]
    xcl68min, xcl68max = get_xpfrac(xg, pdfg, 0.025), get_xpfrac(xg, pdfg, 0.975) 
    xcl95min, xcl95max = get_xpfrac(xg, pdfg, 0.1573), get_xpfrac(xg, pdfg, 0.8427)
    fig_kw={'figsize': figsize}
    fig, ax = plt.subplots(nrows=1, ncols=1, sharex='col', sharey='row', **fig_kw)

    ax.set_xlabel(xlabel)
    ax.set_ylabel('posterior pdf')
    i95 = (xg > xcl95min) & ( xg<=xcl95max)
    ax.fill_between(xg[i95], pdfg[i95], color='slateblue', alpha=0.5, label='95% cl')
    i68 = (xg > xcl68min) & ( xg<=xcl68max)
    ax.fill_between(xg[i68], pdfg[i68], color='slateblue', alpha=0.75, label='68.27% cl')
    ax.plot(xg, pdfg, c='darkslateblue', lw=3)
    ax.legend(loc='best', fontsize=12)
    ax.grid(ls=':', c='gray')

    plt.show()
```

```python
plot_posterior(taug, ptau)
```

### <font color='darkblue'>What we plotted is called the (unnormalized) *posterior distribution*</font>


$$\tilde{\pi}(\tau, t_{\rm wait}\vert \bar{\tau}, \sigma_\tau)=L(t_{\rm wait}\vert \tau)\, p_{\rm G}(\tau\vert \bar{\tau}, \sigma_\tau)$$

$\tilde{\pi}$ in general is not a pdf, even if $L(t_{\rm wait}\vert \tau)$ and $p_{\rm G}(\tau\vert \bar{\tau}, \sigma_\tau)$ are pdfs. 

What should we do to turn it into pdf?

### <font color='darkblue'>Posterior pdf and evidence</font>


We should normalize it, so that integral over all $\tau$ is 1. Posterior pdf is 

$${\pi}(\tau\vert t_{\rm wait}, \bar{\tau}, \sigma_\tau)={\tilde{\pi}(\tau, t_{\rm wait})\over\int \tilde{\pi}(\tau, t_{\rm wait})d\tau }=\frac{1}{p(t_{\rm wait})}\,L(t_{\rm wait}\vert \tau)\, p_{\rm G}(\tau\vert \bar{\tau}, \sigma_\tau)$$

where 

$$p(t_{\rm wait})=\int \tilde{\pi}(\tau, t_{\rm wait})d\tau$$

is called ***the evidence***. 

This is a specific manifestation of what's known as **the Bayes theorem**:

$$p({\rm model\vert data})=\frac{p({\rm data}\vert {\rm model})\,p({\rm model})}{p({\rm data})}$$

We already encountered the Bayes theorem as a consequence of the rules of probability arising from the three axioms of probability (see notebook <tt>02_probability</tt>). 

$$P(A\cdot B) = P(A\vert B)\,P(B)= P(B\vert A)\,P(A), $$

### <font color='darkblue'>Evidence and model comparisons</font>

The evidence - $p({\rm data})$ or $p(t_{\rm twait})$ in our specific example - is probably the trickiest thing to wrap out head around in this approach. It's not easy to understand what the pdf of data measurements in isolation means. 


However, there is no need to overthink it. Its meaning is simply a normalization factor for a given model and data measurements. 


Evidences for two models are used to compare which of the model is a better description of data.
