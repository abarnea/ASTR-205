### <font color='darkblue'>Regression to the mean</font>

First developed by [Francis Galton](https://en.wikipedia.org/wiki/Francis_Galton)  (Charles Darwin's half-cousin) and applied to the data on the heights of parents and their children (Galton 1877, 1886). 

Galton discovered that parents whose average heights (mid-heights) are above the mean tend to have children with heights closer to the mean than their parents, which was called ***regression*** to the mean. 

<br>
<center>
<img width=900 src="http://astro.uchicago.edu/~andrey/classes/a205f20/img/regression_heights.png"></img>

Stated like this, I find it always quite confusing (as do many other people). What this means simply is that the ***slope*** of the relation between the heights of children and the mid-heights of their parents is smaller than 1 (which corresponds to "one-to-one" relation, where heights of children would mostly reflect the height of their parents). 

The slope of the relation is equal to the tangent of the angle the line describing the relation forms with the $x$-axis. 

<br>
<center>
<img width=1100 src="https://qph.fs.quoracdn.net/main-qimg-057a15ea27bc779ee3da8586535ee6b6"></img>

This regression is true in both directions. 

### <font color='darkblue'>Linear regression</font>

The problem of finding the slope and the intercept of such *linear* relations has become known as the *linear regression problem.* 

Linear relation is a relation between quantities $x$ and $y$ which can be described by the linear model constant slope $b$ and intercept $c$: $y_{\rm m}=bx + c$. 

### <font color='darkblue'>Regression: how to estimate the slope of a relation?</font>

Suppose we have two vectors of the same length $n$, $x$ and $y$, with their means $\bar{x}$ and $\bar{y}$. 

Let's define new variables $\bar{x}$, $ \bar{y}$ "centered" on the means (this is often called "pivoting" $x$ and $y$): 

$$\tilde{x} = x - \bar{x};\ \ \ \ \tilde{y} = y - \bar{y}$$

The ***regression process*** is estimating the slope $b_{yx}$ that approximates relation between $\tilde{x}$ and $\tilde{y}$ using a proportionality ("linear") relation: 

$$\tilde{y}_m = b_{yx}\tilde{x},$$

$$\rightarrow\ \ \ y_m = b_{yx} x + \bar{y} - b_{yx}\bar{x} = b_{yx} x + c$$

$$c = \bar{y} - b_{yx}\bar{x}$$

### <font color='darkblue'>The slope estimate: the geometric interpretation</font>

Below you can find formal derivations of the least squares linear regression expression for estimating $b_{xy}$. 

However, before going through the math, it is worth considering how we can estimate $b_{yx}$ *geometrically*. 

The slope $b_{yx}$ is the tangent of the angle formed by the relation with the $x$ axis. Thus, we need a measure of the extent of the data points along $x$ and $y$ directions. The natural choice is standard deviations of $\tilde{x}$ and $\tilde{y}$: $s_x$ and $s_y$. The slope would then simply be the tangent of the angle in the right triangle with the sides $s_x$ and $s_y$: $b_{yx} = s_y/s_x$. 

There is a problem, however. This way of defining the slope means that it can only be positive as standard deviations are always positive. The actual slopes can be negative or positive. Thus, we need a way to assign a sign to the slope. 

The quantity that will have the right sign is something like $\sum_i \tilde{x}_i\tilde{y}_i$ because when $b_{yx}>0$ both $\tilde{x}_i$ and  $\tilde{y}_i$ tend to be either positive or negative. When $b_{yx}<0$ 
$\tilde{x}_i$ and  $\tilde{y}_i$ will tend to have the opposite signs and so result of the sum is either negative or 0. 
The latter will happen when $\tilde{x}_i$ and  $\tilde{y}_i$ values have no correlation with each other and will thus average to 0. 

However, the quantity $\sum_i \tilde{x}_i\tilde{y}_i$ will give us not only sign, but some dependence on the extent of $x$ and $y$. To normalize away this dependence, we can divide it by the product of $\sqrt{\sum_i \tilde{x}_i^2}=\sqrt{(n-1)} s_x$ and
$\sqrt{\sum_i \tilde{y}_i^2}=\sqrt{(n-1)} s_y$ which gives 

$$r = \frac{1}{n-1}\,\frac{\sum_i \tilde{x}_i\tilde{y}_i}{s_x s_y}=\frac{1}{n-1}\,\frac{\sum_{i=1}^n (x_i-\bar{x})(y_i-\bar{y})}{s_x s_y}$$

### <font color='darkblue'>The Pearson correlation coefficient</font>

Named after 19th century British mathematician [Carl Pearson](https://en.wikipedia.org/wiki/Karl_Pearson) who popularized it around 1900, although similar ideas [were introduced by Francis Galton amd Auguste Bravais](https://projecteuclid.org/euclid.ss/1177012580) (the latter also derived the equation for the coefficient). 


$$r = \frac{1}{n-1}\,\frac{\sum_i \tilde{x}_i\tilde{y}_i}{s_x s_y}=\frac{1}{n-1}\,\frac{\sum_{i=1}^n (x_i-\bar{x})(y_i-\bar{y})}{s_x s_y}$$

This is yet another example of mis-attribution of name for a scientific result (aka [the Stigler's law](https://en.wikipedia.org/wiki/Stigler%27s_law_of_eponymy)).

Pearson did also introduce $\chi^2$ test, which we discussed, and which is not named after him. 

Hereafter the Pearson correlation coefficient will be referred to simply as ***the correlation coefficient.***


Given two sequences of numbers, $\vec{x}$ and $\vec{y}$, of length $n$, their sample means, $\bar{x}$ and $\bar{y}$, and variances, $s^2_x$ and $s^2_y$. The sample correlation coefficient is computed as

$$r = \frac{s^2_{xy}}{s_x s_y} = \frac{1}{n-1}\,\sum\limits_{i=0}^{n-1}\,\left(\frac{x_i-\bar{x}}{s_x}\right)\,\left(\frac{y_i-\bar{y}}{s_y}\right) $$


where, as before: 

$$\bar{x} = {1\over n}\,\sum\limits_{i=0}^{n-1}\, x_i; \ \ \ \ \ 
\bar{y} = {1\over n}\,\sum\limits_{i=0}^{n-1}\, y_i$$

$$ s^2_x = {1\over n-1} \sum_{i=0}^{n-1} (x_i - \bar{x})^2;  \ \ \ \ \ 
   s^2_y = {1\over n-1} \sum_{i=0}^{n-1} (y_i - \bar{y})^2 $$

And their standard deviations are computed are  $s_x = \sqrt{s^2_x}$ and $s_y = \sqrt{s^2_y}$.

The correlation coefficient ranges from -1 for perfectly anti-correlated numbers to 1 for perfectly correlated numbers. It is zero, when the two quantities are not correlated. 

**Note.** Just like with the population variance $\sigma^2$ and sample variance $s^2$(estimated for a limited sample of $n$ numbers), different notations are used for the population and sample correlation coefficients: $\rho_{xy}$ is used for the true correlation coefficient for populations $x$ and $y$, while $r_{xy}$ is used to denote the correlation coefficient measured for a limited sample of values from both populations, as defined below. 

The value of the strength of correlation or anti-correlation of values of two different quantities, as illustrated in this figure from Wikipedia (where $r$ is denoted as $\rho$:

## <font color='darkblue'>The linear regression slope
    
We could use this $r$ to define the slope as 
    
    
$$b_{yx}= r\,\frac{s_y}{s_x}=\frac{1}{n-1}\,\frac{\sum_{i=1}^n (x_i-\bar{x})(y_i-\bar{y})}{s_x s_y}\,\frac{s_y}{s_x}=\frac{1}{n-1}\,\frac{\sum_{i=1}^n (x_i-\bar{x})(y_i-\bar{y})}{s_x^2}=\frac{\sum_{i=1}^n (x_i-\bar{x})(y_i-\bar{y})}{\sum_{i=1}^n (x_i-\bar{x})^2}$$


Once we have the slope, we can get the intercept from its definition:

$$c = \bar{y} - b_{yx}\bar{x}.$$

### <font color='darkblue'>Covariance and correlation coefficient</font>

This second term in the expression for the error of the mean is closely related to the notions of covariance and the correlation coefficient. 


***Sample covariance:*** suppose we have two sequences of numbers, $\vec{x}$ and $\vec{y}$, of length $n$, their sample means, $\bar{x}$ and $\bar{y}$, then sample covariance of $x$ and $y$ is defined as: 

$$s^2_{xy} = \frac{1}{n-1}\,\sum\limits_{i=0}^{n-1}\,\left(x_i-\bar{x}\right)\,\left(y_i-\bar{y}\right) = r s_x^2.$$

***Note*** that $x$ and $y$ could be different measurements or just be subsets of the same longer sequence. When $x$ and $y$ are identical, covariance turns into variance.

### <font color='darkblue'>The least square method: estimating the slope by minimizing the sum of squares of the model residuals residual. Formal derivation.</font>

Difference between $y$ data values and model prediction $\tilde{y}_m$:
$$e_i = \tilde{y}_i - \tilde{y}_m.$$

Let's find the value of $b_{xy}$ that minimizes the sum of squares of $e_i$: 

$$E_2 = \frac{1}{n}\,\sum\limits_{i=0}^{n-1}e^2_i = \frac{1}{n}\,\sum\limits_{i=0}^{n-1}(\tilde{y}_i - b_{yx} \tilde{x}_i)^2 = \frac{1}{n}\,\sum\limits_{i=0}^{n-1}(\tilde{y}_i^2 - 2b_{yx} \tilde{x}_i \tilde{y}_i + b^2_{yx}\tilde{x}^2_i)$$

The value of $b_{xy}$ that minimizes $E_2$ corresponds to $\partial E_2/\partial b_{yx} = 0$: 


$$\frac{\partial E_2}{\partial b_{yx}} = \frac{1}{n}\,\sum\limits_{i=0}^{n-1}(2\tilde{x}_i\tilde{y}_i - 2 b_{yx}\tilde{x}^2_i)=\frac{2}{n} \,\sum\limits_{i=0}^{n-1}\tilde{x}_i\tilde{y}_i - 2\frac{b_{yx}}{n}\,\sum\limits_{i=0}^{n-1} \tilde{x}^2_i = 2\frac{(n-1)}{n}\,\left[s^2_{xy} - b_{yx}s^2_x\right]=0,$$

where $s^2_x$ is the sample variance of $x$ values:

$$s^2_x = \frac{1}{n-1}\,\sum\limits_{i=0}^{n-1} \tilde{x}^2_i=\frac{1}{n-1}\,\sum\limits_{i=0}^{n-1} (x_i-\bar{x})^2,$$

and

$$s^2_{xy} = \frac{1}{n-1}\,\sum\limits_{i=0}^{n-1}(x_i-\bar{x})(y_i-\bar{y}).$$




The equality $=0$ is satisfied generally if $s^2_{xy} - b_{yx}s^2_x=0$, which gives: 

$$b_{yx} = \frac{s^2_{xy}}{s^2_x}=\frac{1}{s^2_x(n-1)}\,\sum\limits_{i=0}^{n-1}\,\left(x_i-\bar{x}\right)\,\left(y_i-\bar{y}\right).$$


We can apply the same procedure to determine the model predicting $x$ as a function of $y$:

$$ \tilde{x}_m = b_{xy} \tilde{y}$$

where

$$b_{xy} = \frac{s_{xy}}{s^2_y}=r_{xy}\,\frac{s_x}{s_y}.$$

**Illustration:** consider least squares method equations applied to just 2 points. 

Consider a simple case: $n=2$, when $\bar{x}=(x_0+x_1)/2$ and $\bar{y}=(y_0+y_1)/2$. Then we can show that 

$$y_0 = b_{yx}x_0 + c$$

$$y_1 = b_{yx}x_1 + c$$

Thus, ***the slope and intercept given by the least squares method using just 2 data points $(x_0,y_0)$ and $(x_1,y_1)$ will give a line that goes through these points.*** 

The algebra that shows that the above is true: 

$$b_{yx} = \frac{\sum\limits_{i=0}^{n-1}\,\left(x_i-\bar{x}\right)\,\left(y_i-\bar{y}\right)}{\sum\limits_{i=0}^{n-1}\,\left(x_i-\bar{x}\right)^2}= \frac{\left(x_0-\bar{x}\right)\,\left(y_0-\bar{y}\right) + \left(x_1-\bar{x}\right)\,\left(y_1-\bar{y}\right)}{(x_0-\bar{x})^2 + (x_1-\bar{x})^2}=\frac{(x_0-x_1)\,(y_0-y_1)/4 + (x_1-x_0)\,(y_1-y_0)/4}{(x_0-x_1)^2/4 + (x_0-x_1)^2/4}.$$

Then 

$$b_{yx} =\frac{(y_0-y_1) -(y_1-y_0)}{2(x_0-x_1)} = \frac{y_0-y_1}{x_0-x_1} $$

and with some algebra we can show that 

$$c = \frac{1}{2}(y_0+y_1) - b_{yx}\frac{1}{2}(x_0+x_1)=\frac{1}{2}\left[y_0+y_1 - \frac{y_0-y_1}{x_0-x_1}(x_0+x_1)\right]=\frac{1}{2}\left[\frac{(y_0+y_1)(x_0-x_1) - (y_0-y_1)(x_0+x_1)}{(x_0-x_1)}\right]=\frac{y_1x_0 - y_0x_1}{x_0-x_1}$$

Then 

$$b_{yx}x_0 + c = \frac{y_0-y_1}{x_0-x_1}x_0 + \frac{y_1x_0 - y_0x_1}{x_0-x_1} = y_0$$

$$b_{yx}x_1 + c = \frac{y_0-y_1}{x_0-x_1}x_1 + \frac{y_1x_0 - y_0x_1}{x_0-x_1} = y_1$$


The least squares method is a simplified version of the maximum least squares method. 

### <font color='darkblue'>The weighted least squares method</font>

Data for which we are doing regression often has uncertainties that are specific to each data value (heteroskedastic uncertainties in the statistics jargon). In such a case uncertainties can be taken into account in the least squares method via weights inversely proportional to the square of value's uncertainty (i.e. inversely proportional to its variance). The values with smaller weights (larger uncertainty) will affect result of the regression less than values with smaller uncertainties. 

$$b_{yx, w} = \frac{\sum\limits_{i=0}^{n-1}\,w_i\left(x_i-\bar{x}\right)\,\left(y_i-\bar{y}\right)}{\sum\limits_{i=0}^{n-1}\,w_i\left(x_i-\bar{x}\right)^2}.$$

The intercept $c_w$ is 
$$c_w = \bar{y}_w - b_{yx,w}\bar{x}_w.$$

### <font color='darkblue'>The weighted least squares method: formal derivation</font>


Let's define the "tilde" variables with respect to the *weighted* mean of the data:

$$\bar{x}_w = \frac{\sum\limits_{i=0}^{n-1}w_i x_i}{\sum\limits_{i=0}^{n-1}w_i}; \ \ \ \bar{y}_w = \frac{\sum\limits_{i=0}^{n-1}w_i y_i}{\sum\limits_{i=0}^{n-1}w_i}.$$

$$\tilde{x}_i = x_i - \bar{x}_w;\ \ \ \ \tilde{y}_i = y_i - \bar{y}_w.$$ 

The difference between $y$ data values and model prediction $\tilde{y}_m$:
$$r_i = \tilde{y}_i - \tilde{y}_m=\tilde{y}_i - b_{yx}\tilde{x}_i.$$

Let's find the value of $b_{xy}$ that minimizes the sum of squares of $r_i$, but now with weights $w_i$: 

$$E_2 = \frac{1}{n}\,\sum\limits_{i=0}^{n-1}r^2_i = \frac{1}{n}\,\sum\limits_{i=0}^{n-1}w_i(\tilde{y}_i - b_{yx} \tilde{x}_i)^2 = \frac{1}{n}\,\sum\limits_{i=0}^{n-1}w_i(\tilde{y}_i^2 - 2b_{yx} \tilde{x}_i \tilde{y}_i + b^2_{yx}\tilde{x}^2_i)$$

The value of $b_{xy}$ that minimizes $E_2$ corresponds to $\partial E_2/\partial b_{yx} = 0$: 


$$\frac{\partial E_2}{\partial b_{yx}} = \frac{1}{n}\,\sum\limits_{i=0}^{n-1}w_i(2\tilde{x}_i\tilde{y}_i - 2 b_{yx}\tilde{x}^2_i)=\frac{2}{n} \,\sum\limits_{i=0}^{n-1}w_i\tilde{x}_i\tilde{y}_i - 2\frac{b_{yx}}{n}\,\sum\limits_{i=0}^{n-1} w_i\tilde{x}_i^2 = 0,$$

which implies: 

$$b_{yx, w} = \frac{\sum\limits_{i=0}^{n-1}\,w_i\left(x_i-\bar{x}\right)\,\left(y_i-\bar{y}\right)}{\sum\limits_{i=0}^{n-1}\,w_i\left(x_i-\bar{x}\right)^2}.$$

The intercept $c_w$ is 
$$c_w = \bar{y}_w - b_{yx,w}\bar{x}_w.$$

#### Estimating uncertainties using the bootstrap method

Uncertainties in the slope and intercept estimated with the least squares method can be estimated using the [bootstrap method](https://drive.google.com/file/d/1bqxwc7iDOmvHnypWXCWFZ33LYq1hJHls/view?usp=sharing). However, there would still be no way to incorporate the prior expectation for the parameters, which we know we must do according to the Bayes theorem. 

```python
class LSQ_Regression:
    '''
    Simple (unweighted) least squares regression of data vectors x and y
    '''
    def __init__(self, x=None, y=None):
        self.x = x
        self.y = y
        self.b, self.c = self.slope_and_intercept(x=x, y=y)
            
    def slope_and_intercept(self, x=None, y=None):
        '''
        Estimate slope and intercept using least squares regression
        '''
        if x is None or y is None:
            if self.x is None or self.y is None: 
                raise ValueError("LSQ_Regression error: x or y are not defined!")   
            x, y = self.x, self.y
        nx = len(x)
        assert(nx == len(y)) # this will catch seqiences with unequal size and generate error
        xm = np.mean(x)
        ym = np.mean(y)
        self.b = np.sum((x - xm)*(y - ym)) / np.sum((x-xm)**2)
        self.c = ym - self.b * xm
        return self.b, self.c
        
    def print_slope_and_intercept(self, b=None, c=None):
        '''
        print slope and intercept values 
        '''
        if b is None or c is None: 
            print('b = {:>.4g};  c = {:>.4g}'.format(self.b, self.c))
        else:
            print('b = {:>.4g};  c = {:>.4g}'.format(b, c))
            
```

```python
class WLSQ_Regression(LSQ_Regression):
    '''
    Weighted least squares regression class
    '''
    def __init__(self, x=None, y=None, w=None):
        self.w = w # add weights as an attribute
        super().__init__(x, y) # call init of the parent class 
        
    def wlsq(self, x=None, y=None, w=None):
        '''
        Compute slope and intercept of the linear model using weighted least squares method
        
        Input x, y, w - data x and y values and their weights.
        '''
        if x is None or y is None or w is None:
            if self.x is None or self.y is None or self.w is None: 
                raise ValueError("LinearWLSQ.wlsq error: x or y or w are not defined!")   
            x, y, w = self.x, self.y, self.w
        nx = len(x)
        assert(nx == len(y)) # this will catch seqiences with unequal size and generate error
        xm = np.average(x, weights=w)
        ym = np.average(y, weights=w)

        self.b = np.sum(w*(x - xm)*(y - ym)) / np.sum(w*(x-xm)**2)
        self.c = ym - self.b * xm
        return self.b, self.c

    def wlsq_bootstrap(self, x=None, y=None, w=None, nbs=None): 
        """
        Use bootstrap method to estimate uncertainties on the slope and intercept
        
        make sure that this method can either use input x, y, w or 
        the vectors that were used to initialize the object 
        when an instance of it is created

        if the input nbs number is None, set it internally to 10 or 100 times the size of 
        vector x. 
        """
        if x is None or y is None or w is None:
            if self.x is None or self.y is None or self.w is None: 
                raise ValueError("LinearWLSQ.wlsq error: x or y or w are not defined!")   
            x, y, w = self.x, self.y, self.w
        assert(len(x) == len(y) == len(w))
        nx = len(x)
        # now do bootstrap
        if nbs is None: 
            nbs = nx * 10
        ir = np.random.randint(nx, size=nx*nbs)
        xbs, ybs, wbs = x[ir].reshape(nbs, nx), y[ir].reshape(nbs,nx),  w[ir].reshape(nbs,nx)
        bw, cw = [], []
        for i in range(nbs):
            bd, cd = self.wlsq(xbs[i], ybs[i], wbs[i])
            bw.append(bd); cw.append(cd)

        bbs, ebbs = np.mean(bw), np.std(bw, ddof=1)
        cbs, ecbs = np.mean(cw), np.std(cw,ddof=1)
        covcb = np.cov(bw,cw) # we can also compute the covariance matrix
        print(f'b = {bbs:.3f} +- {ebbs:.4f}; c = {cbs:.3f} +- {ecbs:.4f}')
        print(f'covariance of b and c = {covcb[0,1]:.4f}')
```

### <font color='darkblue'>Example: linear model regression</font>

Estimate the slope $b$ and intercept $c$ for a set of values ${x_i}$, ${y_i}$ and uncertainties in $y$ values (assuming no uncertainty for $x$ values in this case): $\sigma_{y_i}$:

$$y = bx + c,$$

which best describe the correlation between $x$ and $y$.

We will do an experiment on the synthetic (mock) data, where we can control the true value of $b$ and $c$. 

In this experiment we assume that uncertainties for $x_i$ are negligible and only uncertainties for $y_i$ are sizeable

```python
n = 100 # number of measurements to generate
btrue, ctrue, sigma = 3., 3.5, 0.5 # set true values of slope intercept and scale of uncertainties

# we assume that x are randomly distributed in [1,3] interval but their uncertainty is negligible
x  = 1 + 3*np.random.uniform(size=n) # uniformly distribute x values in the range [1,3]
y  = btrue*x + ctrue + np.random.normal(scale=sigma, size=n) # y that will be scattered around linear relation
ey = np.random.normal(loc=sigma, scale=0.2*sigma, size=n)
```

```python
from scipy.special import gammainc

def chi2_pvalue(chi2, k):
    '''
    Compute p-value for chi2 statistic with k degrees of freedom
    '''
    assert(k>0)
    if chi2 / k > 1.0: 
        pvalue = 1 - gammainc(k*0.5, chi2*0.5)
    else:
        pvalue = gammainc(k*0.5, chi2*0.5)
    return pvalue
```

```python
lr = LSQ_Regression(x=x, y=y)
lr.print_slope_and_intercept()

bmodel, cmodel = lr.b, lr.c
```

```python
k = n - 2 # 2 model parameters: slope and interecept 
chi2 = np.sum((y-bmodel*x-cmodel)**2/ey**2) 
chi2_reduced = chi2 / k
print(f'chi2 = {chi2:.3e}; reduced_chi2 = {chi2_reduced:.3f}')

print(f'p-value = {chi2_pvalue(chi2, k):.4g}')
```

```python
# plot data 
plt.figure(figsize=(5,5))
plt.errorbar(x, y, ey, fmt='o', alpha=0.75, color='slateblue', label='data')
plt.xlabel('x'); plt.ylabel('y')
xf = np.linspace(x.min(), x.max(), 10)
plt.plot(xf, bmodel * xf + cmodel, '--', c='g', lw=2, label='a regression model')
plt.plot(xf, btrue * xf + ctrue, '--', c='orangered', label='true model')
plt.legend(loc='upper left', frameon=False)
plt.show()
```

and to the weightest least squares results. For the weights it is customary to addopt $1/\sigma_y^2$:

```python
wlsq = WLSQ_Regression(x=x, y=y, w=1/ey**2)

# compute and print mean WLSQ slope and intercept and their uncertainties
# estimated using bootstrap method
wlsq.wlsq_bootstrap()

bmodel, cmodel = wlsq.b, wlsq.c
```

```python
k = n - 2 # 2 model parameters: slope and interecept 
chi2 = np.sum((y-bmodel*x-cmodel)**2/ey**2) 
chi2_reduced = chi2 / k
print(f'chi2 = {chi2:.3e}; reduced_chi2 = {chi2_reduced:.3f}')

print(f'p-value = {chi2_pvalue(chi2, k):.4g}')
```

```python
# plot data 
plt.figure(figsize=(5,5))
plt.errorbar(x, y, ey, fmt='o', alpha=0.75, color='slateblue', label='data')
plt.xlabel('x'); plt.ylabel('y')
xf = np.linspace(x.min(), x.max(), 10)
plt.plot(xf, bmodel * xf + cmodel, '--', c='g', lw=2, label='a regression model')
plt.plot(xf, btrue * xf + ctrue, '--', c='orangered', label='true model')
plt.legend(loc='upper left', frameon=False)
plt.show()
```

This good agreement between mock data and slope, however, can be fragile if data has intrinsic scatter or its uncertainties are underestimated, as illustrated below. 

```python
n = 100 # number of measurements to generate
btrue, ctrue, sigma = 3., 3.5, 0.5 # set true values of slope intercept and scale of uncertainties

# we assume that x are randomly distributed in [1,3] interval but their uncertainty is negligible
x  = 1 + 3*np.random.uniform(size=n) # uniformly distribute x values in the range [1,3]
y  = btrue*x + ctrue + np.random.normal(scale=sigma, size=n) # y that will be scattered around linear relation
ey = np.random.normal(loc=sigma, scale=0.2*sigma, size=n)

# add extra scatter as intrinsic scatter
s = 1
y = y + np.random.normal(scale=s, size=n)
```

```python
wlsq = WLSQ_Regression(x=x, y=y, w=1/ey**2)

# compute and print mean WLSQ slope and intercept and their uncertainties
# estimated using bootstrap method
wlsq.wlsq_bootstrap()

bmodel, cmodel = wlsq.b, wlsq.c
```

```python
# plot data 
plt.figure(figsize=(5,5))
plt.errorbar(x, y, ey, fmt='o', alpha=0.75, color='slateblue', label='data')
plt.xlabel('x'); plt.ylabel('y')
xf = np.linspace(x.min(), x.max(), 10)
plt.plot(xf, bmodel * xf + cmodel, '--', c='g', lw=2, label='a regression model')
plt.plot(xf, btrue * xf + ctrue, '--', c='orangered', label='true model')
plt.legend(loc='upper left', frameon=False)
plt.show()
```

```python
k = n - 2 # 2 model parameters: slope and interecept 
chi2 = np.sum((y-bmodel*x-cmodel)**2/ey**2) 
chi2_reduced = chi2 / k
print(f'chi2 = {chi2:.3e}; reduced_chi2 = {chi2_reduced:.3f}')

print(f'p-value = {chi2_pvalue(chi2, k):.4g}')
```

The reduced $\chi^2$ in this case is large and p-value is zero, because the least squares model does not account for the intrinsic scatter. In principle, it is possible to include it by adding it "in quadrature" to ey

```python
k = n - 2 # 2 model parameters: slope and interecept 
s = 0.8 # add an estimate of the intrinsic scatter to ey^2 to account for it. 
chi2 = np.sum((y-bmodel*x-cmodel)**2/(ey**2+s**2))
chi2_reduced = chi2 / k
print(f'chi2 = {chi2:.3e}; reduced_chi2 = {chi2_reduced:.3f}')

print(f'p-value = {chi2_pvalue(chi2, k):.4g}')
```
